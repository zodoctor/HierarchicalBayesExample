# A Hierarchical Bayesian Example: Candy!


### What is Hierarchical Bayes?
*Hierarchical Bayesian Analysis* definitely sounds cool, but wtf is it? Well, it's pretty much the usual Bayesian analysis, but with a hierarchy of models in which some models are inputs to other models that eventually trickle down to produce the data. 

### An Example in Words
For example, consider measuring people's heights.  You have a measured height $\hat{h}$ which tells you about the person's *real* height $h$. Inferring the real height from the measured height is the usual Bayesian inference problem:
$$ p(h|\hat{h}) = \frac{p(\hat{h}|h)p(h)}{p(\hat{h})}$$
But what if we consider that one's real height is actually a variable that is dependent on other random variables?  Indeed, one's height isn't a totally unconstrained random variable -- it depends on one's parents' heights and the environment one grew up in. What we've just described is a *hierarchy* of models: One's parents' and environment modulate a child's real height, which then affects the person's measured height.  Let's put this into the math by including the parent and environment information, which we denote as $\mathcal{I}$, in the variables we seek to learn about:
\begin{align}
p(h,\mathcal{I}|\hat{h}) &\propto p(\hat{h}|h,\mathcal{I})p(h,\mathcal{I}) \\
& = p(\hat{h}|h)p(h|\mathcal{I})p(\mathcal{I})
\end{align}
In the second line, we note that $\hat{h}$ only depends on $h$ and use rule of joint probabilities $p(h,\mathcal{I}) = p(h|\mathcal{I})p(\mathcal{I})$. Comparing this to the usual Bayes' rule above, you see that there is a nested structure of probabilities that emerge.  If we want to go even further, we could consider that parent and environment information $\mathcal{I}$ is actually determined by your setting (geography and time), which we could call $\mathcal{S}$. Doing the same thing once again we get:
$$p(h,\mathcal{I},\mathcal{S}|\hat{h}) \propto p(\hat{h}|h)p(h|\mathcal{I})p(\mathcal{I}|\mathcal{S})p(\mathcal{S})$$
You can start to see how each probability term here connects the pieces of the hierarchy to produce the data $\hat{h}$:
$$ \hat{h} \leftarrow h \leftarrow \mathcal{I} \leftarrow \mathcal{S}$$
You can create much more complicated networks of models than this one, where different quantities interact with each other.  Here's an [example](https://www.researchgate.net/figure/Directed-acyclic-graph-for-hierarchical-Bayesian-inference-from-a-training-set-of-SN-Ia_fig1_258369757) of what one might look like. Okay, now let's do a computational example!!

### An Example with M&Ms
Consider eating some M&Ms in a dark room. While the M&Ms all taste the same, they don't all have the same color.  Standard M&Ms come in the colors brown, red, green, orange, yellow, blue, with a total of $N_{\rm colors} = 6$. But since you're in a dark room, your perception of the colors of the M&Ms you're eating isn't perfect, so sometimes you record an M&M as having the wrong color.  The most basic question you might ask is, "What true color Y is this M&M, given that I perceived it as having color X?" We can express our belief in the true color Y given the recorded color X with Bayes' Rule:

$$p(Y|X) = \frac{p(X|Y)p(Y)}{p(X)}$$

$p(X|Y)$ is the likelihood, $p(Y)$ the prior belief on true color Y, and $p(X)$ is called the _evidence_.  First let's look at the likelihood: this tells us how likely the recorded color X is given that the true color was Y. As an example let's suppose that a fraction $f$ of the time you accidentally guess one of the other wrong colors at random. Then,
$$p(X|Y) = \begin{cases} \frac{f}{N_{\rm colors}-1},&X!=Y\\1-f,&X==Y \end{cases}$$
We'll code up this likelihood below.

The evidence term $p(X)$ is the probability of recording the colors you did.  Since the data is fixed, this is a constant and can be ignored in comparing $Y$'s.  Since probabilities must sum to one over all possibilities, i.e. $\sum_Y p(Y|X) = 1$, we get
$$p(X) = \sum_Y p(X|Y)p(Y)$$
This will pop up later....


Finally, the prior $p(Y)$.  Given no other information, one can choose anything they want for the prior.  A common choice is a flat prior over some variables of specific interest.  Here, we'll start by consider each true color Y as having equal probability.  This effectively makes the posterior function proportional to just the likelhood.

Okay, let's code this up!

### The individual M&M color likelihood

```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import matplotlib 
matplotlib.rcParams['text.usetex'] = True
%config InlineBackend.figure_format='retina'
from scipy.stats import multinomial
import seaborn as sns
import pandas as pd
```

```python
class ColorLikelihood:
    def __init__(self,colors,noise_level=0.3):
        """
        Calculate the likelihood of a certain true color
        given the observed color. The noise_level
        is the same as the f parameter described above.
        """
        self.colors = np.array(colors)
        self.ncolors = len(self.colors)
        self.noise_level = noise_level
        
    def likelihood(self,color,data):
        if color==data:
            return 1.-self.noise_level
        else:
            return self.noise_level/(self.ncolors-1)
        
    def true2data(self,true_color):
        if np.random.uniform()>self.noise_level:
            return true_color
        else:
            return np.random.choice(self.colors[self.colors!=true_color])
        
    def draw_data_given_true(self,true_colors):
        true_colors = np.atleast_1d(true_colors)
        return np.array([self.true2data(true_color) for true_color in true_colors])
```

We've given the `ColorLikelihood` class a few helpful methods.  Firstly, it can calculate the likelihood $p(X|Y)$ that we described earlier.  It also has the `draw_data_given_true` function that pushes true M&M colors through the data-generating process to give us *observed* colors so that we can create simulated observations of M&Ms. So with this in hand, we'll genereate some mock observations and then determine the posterior on the color for each of these data points.  Let's assume that each true color of M&M is equally likely to be drawn here, so that the prior is constant over colors and we can just look at the likelihood.

```python
colors = np.array(['red','green','blue','brown','orange','yellow'])
# get observations
nobs = 1000
true_colors = np.random.choice(colors,size=nobs,replace=True)
collk = ColorLikelihood(colors)
data = collk.draw_data_given_true(true_colors)

# put the posteriors in this array of
# dimension number of colors x number of M&Ms
posts = np.array(np.empty((len(colors),nobs),dtype=float))
for ti,dat in enumerate(data):
    for ci,color in enumerate(colors):
        # flat prior means prior = 1./len(colors) 
        posts[ci,ti] = collk.likelihood(color,dat) * (1./len(colors))
```

Okay, we've gotten the individual posteriors for all the M&Ms we've observed! Now let's look at them.  For each M&M let's look at which true color we think is most likely (called the _max a posteriori_ (MaP) estimate of the true color) and compare it to the true color.  How often did we get it right?

```python
map_colors = colors[np.argmax(posts,axis=0)]
# the fraction of the time our MaP guess of the color is the true color
perc_right = 100.*(
    np.sum([
        map_colors[i] == true_colors[i] 
        for i in range(len(true_colors))
    ])
    / len(true_colors)
)

print(f'{perc_right}% of the time the MaP guess of the color is the true color' )
```

Cool, we inferred the true color of each of these M&Ms from our observations, treating each M&M separately, and with a prior that is equal among colors.  We found that ~70% of the time, our maximum a posteriori estimate of the color was indeed the correct color, which corresponds  to the 70% chance that the observed color is the same as the true color.  But what if the bag of M&Ms didn't actually have an equal distribution of colors? Let's create a class for arbritrary priors/populations and test it out!

```python
class CandyPopulation:
    def __init__(self,pcolor_dict):
        self.pcolor_dict = pcolor_dict
        self.colors = list(pcolor_dict.keys())
        self.num2col = {i:self.colors[i] for i in range(len(self.colors))}
        self.probs = list(pcolor_dict.values())
        
    def prior(self,colors):
        if isinstance(colors,str):
            return self.pcolor_dict[colors]
        else:
            return np.array([self.pcolor_dict[color] for color in colors])
        
    def sample(self,n):
        return np.random.choice(self.colors,size=n,replace=True,p=self.probs)

    def p_colors(self,true_colors):
        true_colors = np.atleast_1d(true_colors)
        counts = [np.sum(true_colors==color) for color in self.colors]
        return multinomial.pmf(counts, n=len(counts), p=self.probs)
```

Okay, we've got a class `CandyPopulation` that can (a) give you the prior probability of a certain color with the `prior` function, (b) sample from that prior with `sample` and (c) calculate the probability of getting a certain collection of true colors with `p_colors`. Let's do the same thing we did above, but with an alternate population, and a higher noise level.

```python
pcolor_dict = {'red':0.1,'green':0.3,'brown':0.2,'blue':0.2,'orange':0.19,'yellow':0.01}
colors = np.array(list(pcolor_dict.keys()))
modPop = CandyPopulation(pcolor_dict) # set up our population model
# get observations
nobs = 10000
true_colors = modPop.sample(nobs)
collk = ColorLikelihood(colors,noise_level=0.8)
data = collk.draw_data_given_true(true_colors)
posts_flat = np.array(np.empty((len(colors),nobs),dtype=float)) #posteriors assuming flat prior
posts_mod = np.array(np.empty((len(colors),nobs),dtype=float)) #posteriors assuming modified prior
for ti,true_color in enumerate(true_colors):
    for ci,color in enumerate(colors):
        lk = collk.likelihood(color,data[ti])
        posts_flat[ci,ti] = lk * (1./len(colors))
        posts_mod[ci,ti] = lk * modPop.prior(color)
```

```python
map_colors_flat = colors[np.argmax(posts_flat,axis=0)]
map_colors_mod = colors[np.argmax(posts_mod,axis=0)]

perc_right_flat = 100.*(
    np.sum([
        map_colors_flat[i] == true_colors[i] 
        for i in range(len(true_colors))
    ])
    / len(true_colors)
)

perc_right_mod = 100.*(
    np.sum([
        map_colors_mod[i] == true_colors[i] 
        for i in range(len(true_colors))
    ])
    / len(true_colors)
)

print(f"""
{perc_right_flat:2.1f}% of the time the MaP guess of the color is the true color when
assuming a flat prior on the colors.  
{perc_right_mod:2.1f}% of the time the MaP guess of the color is the true color when
assuming the same prior as the injected population.  
""")
```

Unsurprisingly, we see above that the MaP guesses are better when you use the population prior that corresponds to the actual population. Try playing with the injected population and noise level in the above two cells if you want to get a feel for how things change.  When the injected population is close to equal among colors and the noise level is low, you find that the flat prior does about as well as the "correct" one.  If you start increasing the noise level and changing the injected prior away from flat, you begin to do much better with the true population prior.

Alright, so we've learned that you want to use the population prior when possible.  But what if you don't know the population a priori?  Well, we can use Bayesian inference to infer the population itself!  What we want to know is the population model $\Lambda$, and what we're given is the full set of $N$ M&M observations $\{d_i\}_{i=1}^N$. Enter Bayes' Rule again:
$$p(\Lambda|\{d_i\}) \propto p(\{d_i\}|\Lambda)p(\Lambda)$$
Now note that the term $\propto p(\{d_i\}|\Lambda)$ can be expanded because each $d_i$ really only depends on its corresponding true M&M color $\theta_i$, and $\theta_i$ is only depends on the population model:
$$p(\{d_i\}|\Lambda) =  \int  d\{\theta_i\} p(\{d_i\},\{\theta_i\}|\Lambda) $$
$$ = \int  d\{\theta_i\} p(\{d_i\}|\{\theta_i\},\Lambda)p(\{\theta_i\}|\Lambda) $$
$$ = \prod_i \int  d\theta_i p(d_i|\theta_i)p(\theta_i|\Lambda)$$
In the first line, we explicity show that the $\{\theta_i\}$ are other variables which we have "marginalized out" i.e. averged over. IN the second line, we use the rule of joint probability that $p(A,B) = p(A|B)p(B)$\*. In the final line, we note that individual $\{d_i\}$ are independent of each other (so probabilities just directly multiply) and only depend on $\{\theta_i\}$, so $p(d_i|\theta_i,\Lambda) = p(d_i|\theta_i)$.  Putting this into words, we can calculate $p(\{d_i\}|\Lambda)$ by averaging the likelihood for each individual M&M over all $\theta_i$ drawn from $\Lambda$ and then taking the product over all events.  


\* If there are givens, you always leave them as givens in each term, e.g. $p(A,B|C) = p(A|B,C)p(B|C)$

Okay, let's calculate an example $p(\{d_i\}|\Lambda)$!  Suppose we have two types of bags of M&Ms -- ones from 1997 and ones from now.  The color distribution of M&Ms is different today than it was in 1997, so one might want to ask "given these observed M&Ms I have looked at, how likely is it that I have a bag from 1997 vs now?"  We can answer this question using the likelihood ratio $p(\{d_i\}|\Lambda_{97})/p(\{d_i\}|\Lambda_{\rm now})$.  Let's do that for different numbers of observations!

Instantiate the population models for M&Ms in 1997 vs now. [see this](https://qz.com/918008/the-color-distribution-of-mms-as-determined-by-a-phd-in-statistics/)

```python
mm97_color_dict = {
    'brown':0.3,
    'yellow':0.2,
    'red':0.2,
    'orange':0.1,
    'green':0.1,
    'blue':0.1
}
MM97pop = CandyPopulation(mm_color_dict)
mmnow_color_dict = {
    'brown':0.135,
    'yellow':0.145,
    'red':0.151,
    'orange':0.187,
    'green':0.195,
    'blue':0.187
}
MMnowpop = CandyPopulation(mmnow_color_dict)
```

Create some fake observations of 1997 M&Ms

```python
ntrials = 100
line_colors = ['orange','green','blue']
noise_levels = [0.3,0.5,0.7]
custom_lines = []

# try different noise levels
for ni,noise_level in enumerate(noise_levels):
    BF_lines = []
    
    # do the M&M observation experiment multiple times
    for j in range(ntrials):
        nobs = 1000
        true_colors = MMnowpop.sample(nobs) # synthetic "real" colors from today's M&M population
        collk = ColorLikelihood(list(mmnow_color_dict.keys()),noise_level=noise_level)
        data = collk.draw_data_given_true(true_colors)
        logp_97 = 0.
        logp_now = 0.
        BFs = []
        for i in range(nobs):
            # Here is where we calculate the log of the
            # likelihoods. Since we are using log probabilities
            # we just add them all for all events rather than multiply.
            # The integral in the expressions above is equivalent
            # to the sum shown here since we have discrete categories
            logp_97 += np.log(np.sum(
                [collk.likelihood(color,data[i]) *
                 MM97pop.prior(color) 
                 for color in MM97pop.colors]
            ))
            logp_now += np.log(np.sum(
                [collk.likelihood(color,data[i]) *
                 MMnowpop.prior(color) 
                 for color in MMnowpop.colors]
            ))
            BFs.append(np.exp(logp_97-logp_now))
        plt.plot(range(nobs),BFs,color=line_colors[ni],alpha=0.03)
        BF_lines.append(BFs)
    plt.plot(range(nobs),np.median(BF_lines,axis=0),color=line_colors[ni],linewidth=3)
plt.yscale('log')
plt.xlabel('Number of Observations')
plt.ylabel('likelihood ratio between 97 and now')
plt.ylim(1e-6,10)
```

Wow, so we can see how as we get more observations, we show less and less support for having M&Ms from 97!  With more noise in our observations, it does take longer, but if you observe enough M&Ms, most of the time you get the correct conclusion that you have M&Ms from now!
